*   다변량 그래프 그리기
 * 다변량 그래프: 3개 이상의 변수를 사용한 그래프

* 다변량 그래프가 필요한 이유
 * 3개 이상의 변수를 사용하는 다변량 그래프는 적재 적소에 맞는 그래프 요소를 추가하여 그래프를 표현해야 함

* 다변량 데이터로 다변량 그래프 그리기 - 산점도 그래프
 1. 앞에서 실습한 산점도 그래프에 성별을 새 변수로 추가. 성별은 색상으로 구분. 하지만 Female, Male과 같은 문자열은 산점도 그래프의 색상을 지정하는 값으로 사용할 수 없으므로, 0, 1과 같은 정수를 사용해야 함.
 2. recode_sex 메서드가 반환한 값을 데이터프레임에 추가하기
 3. 테이블당 인원 수 추가하기. 테이블당 인원 수는 점의 크기로 표현. s = 점의 크기, c = 점의 색상



```python
# 여성인 경우 0을 반환하고 남성인 경우 1을 반환하는 함수

def recode_sex(sex):
  if sex == 'Female':
    return 0
  else:
    return 1
```

```python
tips["sex_color"] = tips["sex"].apply(recode_sex)
```

```python
tips.head()
```

```python
# 테이블당 인원 수 산점도 그래프에 추가하기
# Title: Total Bill vs Tip Colored by Sex and Sized by Size
# x: Total Bill
# y: Tip

scatter_plot = plt.figure()
axes1 = scatter_plot.add_subplot(1, 1, 1)
axes1.scatter(
    x = tips["total_bill"],
    y = tips["tip"],
    s = tips["size"] * 10,
    c = tips["sex_color"],
    alpha = 0.5)
axes1.set_title("Total Bill vs Tip Colored by Sex and Sized by Size")
axes1.set_xlabel("Total Bill")
axes1.set_ylabel("Tip")
```

```python
import seaborn as sns

tips = sns.load_dataset("tips")
```

```python
# Total Bill에 대한 히스토그램 그리기
# Title: Total Bill Histogram with Density Plot

ax = plt.subplots()
ax = sns.distplot(tips["total_bill"])
ax.set_title("Total Bill Histogram with Density Plot")
```

```python
# 밀집도 그래프 제외하고 그리기
# Title: Total Bill Histogram
# xlabel: Total Bill
# ylabel: Frequency

ax = plt.subplots()
ax = sns.distplot(tips["total_bill"], kde = False)
ax.set_title("Total Bill Histogram")
ax.set_xlabel("Total Bill")
ax.set_ylabel("Frequency")
```

```python
# 밀집도 그래프만 나타내기
# Title: Total Bill Density
# xlabel: Total Bill
# ylabel: Unit Probability

ax = plt.subplots()
ax = sns.distplot(tips["total_bill"], hist = False)
ax.set_title("Total Bill Density")
ax.set_xlabel("Total BIll")
ax.set_ylabel("Unit Probability")
```

```python
# Title: Total Bill Histogram with Density and Rug Plot
# xlabel: Total Bill

ax = plt.subplots()
ax = sns.distplot(tips["total_bill"], rug = True)
ax.set_title("Total Bill Histogram with Density and Rug Plot")
ax.set_xlabel("Total Bill")
```

* 다양한 종류의 이변량 그래프 그리기
 1. seaborn 라이브러리로 산점도 그래프 그리기: seaborn 라이브러리의 regplot 메서드 이용. 산점도 그래프와 회귀선을 함께 그릴 수 있음. 회귀선을 제거하려면 fit_reg = False
 2. jointplot: 산점도 그래프와 히스토그램을 한 번에 그려줌. x, y 인자에 원하는 열 이름을 지정하고 data 인잣값으로 데이터프레임을 지정
 3. 육각 그래프: 2차원 표면에 육각형으로 데이터를 쌓아 표현하는 그래프. 산점도 그래프의 데이터를 구분하기 쉽게 그리고 싶을 때 사용. jointplot 메서드를 그대로 사용하지만 kind 인잣값을 hex로 지정

 5. 바 그래프 그리기: 지정한 변수의 평균의 계산하여 그리기
 6. 박스 그래프 그리기: 최솟값, 중간값, 3분위수, 최댓값, 이상치 등 다양한 통계량을 한 번에 표현
 7. 박스 그래프는 데이터 분산이 모호하게 표현됨. 이런 경우 violinplot을 이용한 바이올린 그래프를 사용
 8. 관계 그래프 그리기: 지금까지 설명한 그래프를 종하바 그래프. pairplot 메서드에 데이터프레임을 넣음
 9. 관계 그래프는 중복된 정보가 표현됨. 중복된 그래프가 그려지는 위치를 직접 지정하여 원하는 그래프로 교체하면 됨. map_upper 메서드는 대각선을 기준으로 위쪽에 그릴 그래프를 지정. map_lower 메서드는 대각선을 기준으로 아래쪽에 그릴 그래프를 지정. map_diag는 대각선을 중심으로 그래프를 그림.

```python
# Title: Scatterplot of Total Bill and Tip
# xlabel: Total Bill
# ylabel: Tip

ax = plt.subplots()
ax = sns.regplot(x = "total_bill", y = "tip", data = tips)
ax.set_title("Scatterplot of Total Bill and Tip")
ax.set_xlabel("Total Bill")
ax.set_ylabel("Tip")
```

```python
# Title: Scatterplot of Total Bill and Tip
# xlabel: Total Bill
# ylabel: Tip

ax = plt.subplots()
ax = sns.regplot(x = "total_bill", y = "tip", data = tips, fit_reg = False)
ax.set_title("Scatterplot of Total Bill and Tip")
ax.set_xlabel("Total Bill")
ax.set_ylabel("Tip")
```

```python
# xlabel: Total Bill
# ylabel: Tip
# title: Joint Plot of Total Bill and Tip
# fontsize = 10
# y = 1.03

joint = sns.jointplot(x = "total_bill", y = "tip", data = tips)
joint.set_axis_labels(xlabel = "Total Bill", ylabel = "Tip")
joint.fig.suptitle("Joint Plot of Total Bill and Tip", fontsize = 10, y = 1.03)
```

```python
# xlabel: Total Bill
# ylabel: Tip
# title: Hexbin Joint Plot of Total Bill and Tip
# fontsize = 10
# y = 1.03

hexbin = sns.jointplot(x = "total_bill", y = "tip", data = tips, kind = "hex")
hexbin.set_axis_labels(xlabel = "Total Bill", ylabel = "Tip")
hexbin.fig.suptitle("Hexbin Joint Plot of Total Bill and Tip", fontsize = 10, y = 1.03)
```

```python
# xlabel = Total Bill
# ylabel = Tip
# Title = Karnel Density Plot of Total Bill and Tip

ax = plt.subplots()
ax = sns.kdeplot(x = tips["total_bill"],
                 y = tips["tip"],
                 shade = True)
ax.set_title("Karnel Density Plot of Total Bill and Tip")
ax.set_xlabel("Total Bill")
ax.set_ylabel("Tip")
```

```python
# title = Bar plot of averae total bill for time of day
# xlabel = Time of Day
# ylabel = Average total bill

ax = plt.subplots()
ax = sns.barplot(x = "time", y = "total_bill", data = tips)
ax.set_title("Bar plot of average total bill for time of day")
ax.set_xlabel("Time of Day")
ax.set_ylabel("Average total bill")
```

```python
# Title = Boxplot of total bill by time of day
# xlabel = Time of day
# ylabel = Total Bill

ax = plt.subplots()
ax = sns.boxplot(x = "time", y = "total_bill", data = tips)
ax.set_title("Boxplot of total bill by time of day")
ax.set_xlabel("Time of day")
ax.set_ylabel("Total Bill")
```

```python
# Title = Violin plot of total bill by time of day
# xlabel = Time of day
# ylabel = Total Bill

ax = plt.subplots()
ax = sns.violinplot(x = "time", y = "total_bill", data = tips)
ax.set_title("Violin plot of total bill by time of day")
ax.set_xlabel("Time of day")
ax.set_ylabel("Total Bill")
```

**04-4 데이터프레임과 시리즈로 그래프 그리기**
 * 데이터프레임과 시리즈로 그래프 그리기
  1. 시리즈에 있는 plot 속성에 정의된 hist 메서드를 사용하면 히스토그램을 그릴 수 있음.
  2. 투명도를 조절하려면 hist 메서드의 alpha, bins, ax 인자를 이용
  3. kide = 밀집도, scatter = 산점도 그래프, hexbin = 육각 그래프
  4. 육각 그래프의 크기는 gridsize 인자를 사용하여 변경할 수 있음.
  5. box 메서드를 사용하여 박스 그래프를 그릴 수 있음.

```python
ax = plt.subplots()
ax = tips["total_bill"].plot.hist()
```

```python
# 투명도 = 0.5, x축의 간격 = 20
fig, ax = plt.subplots()
ax = tips[["total_bill", "tip"]].plot.hist(alpha = 0.5, bins = 20, ax = ax)
```

```python
# x = total bill, y = tip의 산점도 그래프 그리기
fig, ax = plt.subplots()
ax = tips.plot.scatter(x = "total_bill", y = "tip", ax = ax)
```

```python
# x = total bill, y = tip의 육각 그래프 그리기
fig, ax = plt.subplots()
ax = tips.plot.hexbin(x = "total_bill", y = "tip", ax = ax)
```

```python
# 육각 그래프의 사이즈 변경하기
fig, ax = plt.subplots()
ax = tips.plot.hexbin(x = "total_bill", y = "tip", gridsize = 10, ax = ax)
```

**04-5 seaborn 라이브러리로 그래프 스타일 설정하기**
 * seaborn 라이브러리로 만든 그래프는 모두 기본 스타일. set_style 메서드를 사용하여 그래프의 스타일을 바꿀 수 있음. 메서드를 실행한 이후의 그래프가 모두 저장한 스타일로 바뀜.

 * 그래프에 스타일 적용하기
  2. whitegrid 스타일을 생성하여 그래프의 배경에 가로줄 만들기
  3. for문을 이용하여 모든 스타일을 하나씩 적용한 그래프

```python
fig, ax = plt.subplots()
ax = sns.violinplot(x = "time", y = "total_bill", hue = "sex", data = tips, split = True)
```

```python
sns.set_style("whitegrid")
fig, ax = plt.subplots()
ax = sns.violinplot(x = "time", y = "total_bill", hue = "sex", data = tips, split = True)
```

```python
fig = plt.figure()

seaborn_styles = ['darkgrid', 'whitegrid', 'dark', 'white', 'ticks']

for idx, style in enumerate(seaborn_styles):
  plot_position = idx + 1
  with sns.axes_style(style):
    ax = fig.add_subplot(2, 3, plot_position)
    violin = sns.violinplot(x = "time", y = "total_bill", data = tips, ax = ax)
    violin.set_title(style)

fig.tight_layout()
```
