4.   데이터 핸들링 - 판다스



판다스:
 *   파이썬에서 데이터 처리를 위해 존재하는 가장 인기 있는 라이브러리.
 *   행과 열로 이뤄진 2차원 데이터를 효율적으로 가공/처리할 수 있는 다양하고 훌륭한 기능 제공.
 * 넘파이보다 훨씬 유연하고 편리하게 데이터 핸들링을 가능하게 해줌.

DataFrame: 여러 개의 행과 열로 이뤄진 2차원 데이터를 담는 데이터 구조체.

Index: 개별 데이터를 고유하게 식별하는 Key값

Series와 DataFrame은 모두 Index를 key 값으로 가지고 있지만, Series는 칼럼이 하나뿐인 데이터 구조체이고, DataFrame은 칼럼이 여러 개인 데이터 구조체다.

**판다스 시작 - 파일을 DataFrame으로 로딩, 기본 API**

*   read_csv(): csv 파일 포맷 변환을 위한 API. sep에 구분 문자를 입력하면 어떤 필드 구분 문자 기반의 파일 포맷도 DataFrame으로 변환 가능. 호출 시 파일명 인자로 들어온 파일을 로딩해 DataFrame 객체로 변환. 별다른 파라미터 지정이 없으면 파일의 맨 처음 로우를 갈럼명으로 인지하고 칼럼으로 변환함.
*   read_table(): 디폴트 필드 구분 문자는 탭 문자
*   read_fwf(): 고정 길이 기반의 칼럼 포맷을 DataFrame으로 로딩하기 위한 API.
*   filepath: 로드하려는 데이터 파일의 경로를 포함한 파일명을 입력하면 됨.
*   DataFrame.head(): DataFrame의 맨 앞에 있는 N개의 로우를 반환함. (Default는 5개)
*   shape: DataFrame의 행과 열 크기를 알아보는 가장 좋은 방법법
*   info(): 메서드를 통해서 총 데이터 건수와 데이터 타입, Null 건수를 알 수 있음.
*   describe(): 오직 숫자형(int, float 등) 칼럼의 분포도만 조사하여 자동으로 object 타입의 칼럼은 출력에서 제외시킴.
*   count: Not Null인 데이터 건수
*   mean: 전체 데이터의 평균값
*   sd: 전체 데이터의 표준편차
*   min: 전체 데이터 중 최솟값
*   max: 전체 데이터 중 최댓값
*   25%, 50%, 75%: 25 percentile, 50 percentile, 75 percentile 값을 의미
*   value_counts(): 지정된 칼럼의 데이터값 건수를 반환. 데이터의 분포도를 확인하는 데 매우 유용한 함수. 칼럼 값별 데이터 건수를 반환하므로 고유 칼럼 값을 식별자로 사용할 수 있음.
*   Series: Index와 단 하나의 칼람으로 구성된 데이터 세트
*   모든 Series와 DataFrame은 인덱스를 반드시 가짐. value_counts는 Series 객체에만 정의되어 있고, DataFrame에서는 갖고 있지 않음.
*   Index는 단순히 순차 값과 같은 의미 없는 식별자만 할당하는 것이 아니라 고유성이 보장된다면 의미 있는 데이터값 할당도 가능. Index는 숫자형뿐만 아니라 문자열도 가능. 단, 모든 Index는 고유성이 보장돼야 함.

**DataFrame과 리스트, 딕셔너리, 넘파이 ndarray 상호 변환**
*   DataFrame은 파이썬의 리스트, 딕셔너리 그리고 넘파이 ndarray 등 다양한 데이터로부터 생성될 수 있고, DataFrame은 반대로 파이썬의 리스트, 딕셔너리 그리고 넘파이 ndarray 등으로 변환될 수 있음. DataFrame과 넘파이 ndarray 상호 간의 변환은 매우 빈번하게 발생.

넘파이 ndarray, 리스트, 딕셔너리를 DataFrame으로 변환하기
*   DataFrame은 리스트와 넘파이 ndarray와 다르게 칼럼명을 가지고 있음. 리스트와 넘파이 ndarray보다 상대적으로 편하게 데이터 핸들링이 가능. 

```python
import numpy as np

col_name1 = ['col1']
list1 = [1, 2, 3]
array1 = np.array(list1)
print('array1 shape:', array1.shape)

# 리스트를 이용해 DataFrame 생성.
df_list1 = pd.DataFrame(list1, columns = col_name1)
print('1차원 리스트로 만든 DataFrame:\n', df_list1)

# 넘파이 ndarray를 이용해 DataFrame 생성.
df_array1 = pd.DataFrame(array1, columns = col_name1)
print('1차원 ndarray로 만든 DataFrame:\n', df_array1)
```

```python
# 3개의 칼럼명이 필요함.
col_name2 = ['col1', 'col2', 'col3']

# 2행x3열 형태의 리스트와 ndarray 생성한 뒤 이를 DataFrame으로 변환.
list2 = [[1, 2, 3],
         [11, 12, 13]]
array2 = np.array(list2)
print('array2 shape: ', array2.shape)
df_list2 = pd.DataFrame(list2, columns = col_name2)
print('2차원 리스트로 만든 DataFrame:\n', df_list2)
df_array2 = pd.DataFrame(array2, columns = col_name2 )
print('2차원 ndarray로 만든 DataFrame:\n', df_array2)
```

```python
# Key는 문자열 칼럼명으로 매핑, Value는 리스트 형(또는 ndarray) 칼럼 데이터로 매핑
dict = {'col1': [1, 11], 'col2': [2, 12], 'col3': [3, 13]}
df_dict = pd.DataFrame(dict)
print('딕셔너리로 만든 DataFrame:\n', df_dict)
```

DataFrame을 넘파이 ndarray, 리스트, 딕셔너리로 변환하기
*   데이터 핸들링은 DataFrame을 이용하더라도 머신러닝 패키지의 입력 인자 등에 적용하기 위해 다시 넘파이 ndarray로 변환하는 경우가 빈번하게 발생함.
*   DataFrame -> ndarray 변환: values를 이용
*   DataFrame -> 리스트로 변환: values로 얻은 ndarray에 tolist()를 호출
*   DataFrame -> 딕셔너리로 변환: DataFrame 객체의 to_dict() 메서드를 호출. 인자로 list를 입력하면 딕셔너리의 값이 리스트형으로 변환됨.

```python
# DataFrame을 ndarray로 변환
array3 = df_dict.values
print('df_dict.values 타입: ', type(array3), 'df_dictvalues shape: ', array3.shape)
print(array3)
```

```python
# DataFrame을 리스트로 변환
list3 = df_dict.values.tolist()
print('df_dict.values.tolist() 타입: ', type(list3))
print(list3)

# DataFrame을 딕셔너리로 변환
dict3 = df_dict.to_dict('list')
print('\n df_dict.to_dict() 타입: ', type(dict3))
print(dict3)
```

**DataFrame 데이터 삭제**
*   drop() 메서드 이용.
*   axis 0은 로우 방향 축, axis 1은 칼럼 방향 축. axis = 1을 입력하면 칼럼 축 방향으로 드롭. axis = 0을 입력하면 로우 축 방향으로 드롭
*   inplace: 디폴트값이 False 이므로 자기 자신의 DataFrame 데이터는 삭제하지 않으며, 삭제된 결과 DataFrame을 반환함. inplace = True로 설정하면 자신의 DataFrame의 데이터를 삭제함.

<drop() 메서드에서 axis와 inplace 인자를 적용해 DataFrame을 변경하는 방식>
1.   axis: DataFrame의 로우를 삭제할 때는 axis = 0, 칼럼을 삭제할 때는 axis = 1으로 설정.
2.   원본 DataFrame은 유지하고 드롭된 DataFrame을 새롭게 객체 변수로 받고 싶다면 inplace=False로 설정(디폴트 값이 False임).
3.   원본 DataFrame에 드롭된 결과를 적용할 경우에는 inplace=True를 적용.
4.   원본 DataFrame에서 드롭된 DataFrame을 다시 원본 DataFrame 객체 변수로 할당하면 원본 DataFrame에서 드롭된 결과를 적용할 경우와 같음.(단, 기존 원본 DataFrame 객체 변수는 메모리에서 추후 제거됨).

**Index 객체**

*   DataFrame, Series의 레코드를 고유하게 식별하는 객체.
*   식별성 데이터를 1차원 array로 가지고 있음. ndarray와 유사하게 단일 값 반환 및 슬라이싱도 가능함.
*   한 번 만들어진 DataFrame 및 Series의 Index 객체는 함부로 변경할 수 없음.
*   Series 객체는 Index 객체를 포함하지만 Series 객체에 연산 함수를 적용할 때 Index는 연산에서 제외됨.

**데이터 셀렉션 및 필터링**
*   넘파이의 경우 '[]' 연산자 내 단일 값 추출, 슬라이싱, 팬시 인덱싱, 불린 인덱싱을 통해 데이터를 추출함.
*   판다스의 경우 ix[], iloc[], loc[] 연산자를 통해 동일한 작업을 수행함.

DataFrame의 [] 연산자
*   넘파이에서 [] 연산자는 행의 위치, 열의 위치, 슬라이싱 범위 등을 지정해 데이터를 가져올 수 있었음
*   DataFrame 바로 뒤에 있는 '[]' 안에 들어갈 수 있는 것은 칼럼 명 문자, 또는 인덱스로 변환 가능한 표현식임.
*   DataFrame 뒤에 있는 []는 칼럼만 지정할 수 있는 '칼럼 지정 연산자'로 이해하는 게 가장 좋음.
*   판다스의 인덱스 형태로 변환 가능한 표현식은 [] 내에 입력할 수 있음.
*   불린 인덱싱 표현도 가능.
*   DataFrame 바로 뒤의 [] 연산자는 넘파이의 []나 Series의 []와 다름
*   DataFrame 바로 뒤의 [] 내 입력 값은 칼럼명(또는 칼럼의 리스트)을 지정해 칼럼 지정 연산에 사용하거나 불린 인덱스 용도로만 사용해야 함.
*   DataFrame[0:2]와 같은 슬라이싱 연산으로 데이터를 추출하는 방법은 사용하지 않는게 좋음.

명칭 기반 인덱싱과 위치 기반 인덱싱의 구분

*   위치(Position) 기반 인덱싱: 0을 출발점으로 하는 가로축, 세로축 좌표 기반의 행과 열 위치를 기반으로 데이터를 지정함.

DataFrame loc[] 연산자
*   loc[]는 명칭 기반으로 데이터를 추출함.
*   행 위치에는 DataFrame index 값을, 열 위치에는 칼럼 명을 입력해 줌.
*   loc[]에 슬라이싱 기호를 적용하면 종료 값-1이 아니라 종료 값까지 포함하는 것을 의미함.

<ix, iloc, loc의 문제점과 주의 할 점>
1.   명칭 기반 인덱싱과 위치 기반 인덱싱의 차이를 이해해야 함. DataFrame의 인덱스나 칼럼명으로 데이터에 접근하는 것은 명칭 기반 인덱싱. 0부터 시작하는 행, 열의 위치 좌표에만 의존하는 것이 위치 기반 인덱싱.
2.   ix[]는 명칭 기반 인덱싱과 위치 기반 인덱싱을 모두 적용 가능. 
3.   iloc[]는 위치 기반 인덱싱만 가능함. =
4.   loc[]는 명칭 기반 인덱싱만 가능함.
5.   명칭 기반 인덱싱에서 슬라이싱을 '시작점:종료점'으로 지정할 때 시작점에서 종료점을 포함한 위치에 있는 데이터를 반환함.

**정렬, Aggregation 함수, GroupBy 적용**
DataFrame, Series의 정렬 - sort_values()
*   sort_values(): DataFrame과 Series 정렬할 때 사용. 주요 입력 파라미터는 by, ascending, inplace.
*   by: 특정 칼럼을 입력하면 해당 칼럼으로 정렬을 수행.
*   ascending = True로 설정하면 오름차순으로 정렬, ascending = False로 설정하면 내림차순으로 정렬. (기본은 ascending = True)
*   inplace = False로 설정하면 sort_values()를 호출한 DataFrame은 그대로 유지하며 정렬된 DataFrame을 반환함. inplace = True로 설정하면 호출한 DataFrame의 정렬 결과를 그대로 적용함.(기본은 inplace = False)

```python
titanic_sorted = titanic_df.sort_values(by = ['Name'])
titanic_sorted.head(3)
```

```python
titanic_sorted = titanic_df.sort_values(by = ['Pclass', 'Name'], ascending = False)
titanic_sorted.head(3)
```

*   isna(): NaN 여부를 확인하는 API
*   filna(): NaN 값을 다른 값으로 대체하는 API

apply lambda 식으로 데이터 가공
*   apply 함수에 lambda 식을 결합해 DataFrame이나 Series의 레코드별로 데이터를 가공하는 기능을 제공함.
*   lambda 식: 파이썬에서 함수형 프로그래밍을 지원하기 위해 만들음.
*   map(): 여러 개의 값을 입력 인자로 사용해야 할 경우 사용
*   if else를 사용할 때 if 식보다 반환 값을 먼저 기술해야 함.
*   if, else만 지원하고 if, else if, else와 같이 else if는 지원하지 않음. else if를 이용하기 위해서는 else 절을 ()로 내포해 () 내에서 다시 if else를 적용해 사용함.
*   switch case 문의 경우는 else를 계속 내포해서 쓰기에는 부담스러움. -> 아예 별도의 함수의 만드는게 나음.

```python
def get_square(a):
  return a**2

print('3의 제곱은:', get_square(3))
```

```python
lambda_square = lambda x : x**2
print('3의 제곱은: ', lambda_square(3))
```

```python
a = [1, 2, 3]
squares = map(lambda x : x**2, a)
list(squares)
```

```python
titanic_df['Name_len'] = titanic_df['Name'].apply(lambda x : len(x))
titanic_df[['Name', 'Name_len']].head(3)
```

```python
titanic_df['Child_Adult'] = titanic_df['Age'].apply(lambda x : 'Child' if x <= 15 else 'Adult')
titanic_df[['Age', 'Child_Adult']].head(8)
```

```python
titanic_df['Age_cat'] = titanic_df['Age'].apply(lambda x : 'Child' if x <= 15 else ('Adult' if x<= 60 else 'Elderly'))
titanic_df['Age_cat'].value_counts()
```

```python
# 나이에 따라 세분화된 분류를 수행하는 함수 생성.
def get_category(age):
  cat = ''
  if age <= 5: cat = 'Baby'
  elif age <= 12: cat = 'Child'
  elif age <= 18: cat = 'Teenager'
  elif age <= 25: cat = 'Student'
  elif age <= 35: cat = 'Young Adult'
  elif age <= 60: cat = 'Adult'
  else: cat = 'Elderly'

  return cat

# lambda 식에 위에서 생성한 get_category() 함수를 반환값으로 지정.
# get_category(X)는 입력값으로 'Age' 칼럼 값을 받아서 해당하는 cat 반환
titanic_df['Age_cat'] = titanic_df['Age'].apply(lambda x : get_category(x))
titanic_df[['Age', 'Age_cat']].head()
```
